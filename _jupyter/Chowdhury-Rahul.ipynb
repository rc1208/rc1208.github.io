```python
#make a list of all the 100 files which have edge list of the 100 universities by excluding the metadata files
files = [fn for fn in glob.glob('facebook100txt/*') 
         if not os.path.basename(fn).endswith('_attr.txt')]
len(files)
```

```python
ratio = {} #y-axis dict
mean_k_u_list = {} #x-axis dict
for f in files: #read each file
    G = nx.Graph()
    G = nx.read_edgelist(f)
    deg_list = dict(G.degree())
    numerator = 0 #Average squared degree
    denominator = 2*len(G.edges()) #Average degree
    for key,value in deg_list.items():
        numerator = numerator + value*value
        
    k_v = numerator/denominator
    k_u = 2 *len(G.edges)/len(G.nodes)
    ratio[f] = k_v/k_u  
    mean_k_u_list[f] = k_u
    
```

```python
Reed_x = mean_k_u_list['facebook100txt/Reed98.txt']
Reed_y = ratio['facebook100txt/Reed98.txt']
Bucknell_x = mean_k_u_list['facebook100txt/Bucknell39.txt']
Bucknell_y = ratio['facebook100txt/Bucknell39.txt']
Mississipi_x = mean_k_u_list['facebook100txt/Mississippi66.txt']
Mississipi_y = ratio['facebook100txt/Mississippi66.txt']
Virginia_x = mean_k_u_list['facebook100txt/Virginia63.txt']
Virginia_y = ratio['facebook100txt/Virginia63.txt']
Uc_Berkeley_x = mean_k_u_list['facebook100txt/Berkeley13.txt']
Uc_Berkeley_y = ratio['facebook100txt/Berkeley13.txt']
```

```python
def temporal_graph(x_data,y_data,xlabel,ylabel,title):
    # Create the plot object
    _, ax = plt.subplots()
    ax.grid()
    ax.set_ylabel(ylabel)
    ax.set_xlabel(xlabel)
    ax.set_title(title)
    # Plot the data, set the size (s), color and transparency (alpha)
    # of the points
    ax.scatter(x_data, y_data, color = '#539caf', alpha = 0.75)
    
    plt.axhline(y=1, color='b', linestyle='-', label = "line")
    R, = ax.plot(Reed_x,Reed_y, "bo")
    B, = ax.plot(Bucknell_x,Bucknell_y, "or")
    M, = ax.plot(Mississipi_x,Mississipi_y, "yo")
    V, = ax.plot(Virginia_x,Virginia_y, "ko")
    U, = ax.plot(Uc_Berkeley_x,Uc_Berkeley_y, "go")
    plt.legend([R,B,M,V,U],['Reed','Bucknell','Mississipi','Virginia','UC Berkeley'])
   
```

```python
x_data = []
y_data = []
for key,value in ratio.items():
    x_data.append(mean_k_u_list[key])
    y_data.append(ratio[key])
```

### Q6 b) The friendship paradox graph

```python
temporal_graph(x_data,y_data,"mean k_u","mean k_v/ mean k_u", "Friendship Paradox")
```

### Q 6 d) Finding diameter of largest component of each network $(l_{max})$ ###

```python
diameter = {}
for f in files: #read each file
    G = nx.Graph()
    G = nx.read_edgelist(f)
    giant = max(nx.connected_component_subgraphs(G), key=len)
    d = nx.diameter(giant, e=None)
    print(f, d)
    diameter[f] = d 
```
